```python
def triangle(length):
    Y = []
    for x in range(length):
        x = x % 6
        if x <= 3:
            Y.append(x)
        else:
            Y.append(6-x)
    return Y

def sin(length):
    X = np.arange(length)
    Y = (np.sin(X * 0.1) * 10).astype(np.int) + 10
    return Y
```

```python
Y = triangle(100)
Y = sin(100)
plt.plot(Y)
```
